# Data preprocessing

```python
import pyreadr

# Read data from rds file
data = pyreadr.read_r('data/f1_dat.rds')
# Extract the df
df = data[None]
# Drop irrelevant columns 
df.drop(['round', 'circuit', 'weather_type', 'circuit_type', 'status'], axis=1, inplace=True)
# Select years 2017-2021
df = df[df['year'] >= 2017]
df.reset_index(drop=True, inplace=True)
# Unify constructor names
df = df.replace({'force_india': 'aston_martin',
                 'racing_point': 'aston_martin',
                 'toro_rosso': 'alphatauri',
                 'renault': 'alpine',
                 'sauber': 'alfa'})
# Change race positions range to 0-19
df['position'] -= 1
# Export data
df.to_csv('data/f1_results.csv', index=False)

df
```
